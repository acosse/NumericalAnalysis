```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100)



def Chebyshev(n, x):
''' La fonction renvoie les coefficients du polynome de Chebyshev ainsi
    que les zeros de ce polynome'''

  p = np.cos(n*np.arccos(x))
  k = np.arange(1, n+1)
  xk = np.cos(((2*k - 1)/2/n) * np.pi)

  return p, xk


plt.plot(x, Chebyshev(10,x)[0])
plt.scatter(Chebyshev(10,x)[1], np.zeros((len(xk),1)), c='r')
plt.show()

```

```python

x = np.linspace(-1,1,100)

# defining Runge function
def Runge(x):

  f = 1./(1+25*x**2)

  return f

n = 20
xi = np.cos(((2*np.arange(1,n+1)-1)/2/n) * np.pi)

f1 = Runge(x)
# calculating Chebyshev interpolation points
fi = Runge(xi)
# calculating Chebyshev interpolation polynomial
p_Cheby = np.poly1d(np.polyfit(xi, fi, n-1))

# calculating the interpolation polynomial at equispaced points
x_equi = np.linspace(-1,1,n)
f_equi = Runge(x_equi)

p_equi = np.poly1d(np.polyfit(x_equi, f_equi, n-1))



plt.plot(x, f1)
plt.plot(x, p_Cheby(x))
#plt.plot(x, p_equi(x))

plt.show()
```
