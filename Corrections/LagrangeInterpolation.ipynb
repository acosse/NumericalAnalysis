```python

# importing numpy and pyplot librairies
import numpy as np
import matplotlib.pyplot as plt


# Part I: Lagrange interpolation

# Part 1. we start by defining the functions lagrange and interpolating_polynomial that
# respectively compute (1) the k^th Lagrange polynomial and (2) the interpolation polynomial given a set of (x, y) pairs


def lagrange(k,x,z):
  ''' function should return the kth Lagrange polynomial l_k(z) at the point (or set of points z)
      for the set of interpolation nodes x = [x_0, x_1, ..., x_n]'''

  lagrange = np.ones((len(z), 1))

  for j in range(len(z)):

    for i in range(len(x)):
            # looping over all interpolation points
      if i!= k:

        # calculation of Lagrange polynomial

        lagrange[j]  = lagrange[j]* (z[j] - x[i])/(x[k] - x[i])

  return lagrange

def interpolating_polynomial(x, y, z):
  ''' function should return the interpolating polynomial of degree n defined on the n+1 points
      (x_0, y_0) (x_1, y_1), ... (x_n, y_n) p_n(x) = sum_{k=0}^n y_k l_k(z) evaluated at the
      point, or set of points z'''
  polynomial = 0

  for j in range(len(y)):

      lagrange_j = lagrange(j,x,z)

      polynomial = polynomial + lagrange_j*y[j]


  return polynomial


# Applying the functions lagrange and interpolation polynomial

x = np.linspace(0,4,5)

y = np.sqrt(np.pi*x)*np.cos(np.pi*x)

plt.scatter(x, y, c='r')
plt.title('Interpolation points')
plt.show()
```

```python

z = np.linspace(0,4,100)

poly_interp = interpolating_polynomial(x, y, z)

plt.plot(z, poly_interp)
plt.scatter(x, y, c='r')
plt.title('Interpolation polynomial')
plt.show()


```

```python
# Representing each of the Lagrange polynomials

z = np.linspace(0, 4, 100)

plt.title('Lagrange Polynomials')

for k in range(len(x)):

  plt.plot(z, lagrange(k, x, z))
```

```python
# Part II: An alternative consist in using the functions polyfit

'''The function polyfit returns the vector of coefficients of the
interpolating polynomial p_n of degree n (len(x)-1 in this case) defined on
the n=1 points (x_0, y_0), ... (x_n, y_n)'''

interpolation_poly = np.polyfit(x, y, len(x)-1)

# to evaluate the polynomial, we can use the lines

p = np.poly1d(interpolation_poly)
z = np.linspace(0,4,100)


plt.title('interpolation polynomial returned by polyfit')
plt.scatter(x, y, c='r')
plt.plot(z, p(z))
plt.show()



```
